```python
#import the libraries
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# load dataset

df=pd.read_csv(r'C:\Users\BlackDevil\Downloads\data\epl_20.csv')
df
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isna().sum()
```

```python
# create new column
df['minpermatch']=(df['Mins']/df['Matches']).astype(int)
df['Goalspermatch']=(df['Goals']/df['Matches']).astype(float)
df.head()
```

```python
# total goals
total_goals=df['Goals'].sum()
print(total_goals)
```

```python
# penalty goals
penalty_goals=df['Penalty_Goals'].sum()
print(penalty_goals)
```

```python
# penalty attempts
total_Penaltyattempts=df['Penalty_Attempted'].sum()
print(total_Penaltyattempts)
```

```python
# total assist
assist=df['Assists'].sum()
print(assist)
```

```python
# pie chart for penalites missed vs scored

plt.figure()
pl_not_scored=df['Penalty_Attempted'].sum() - penalty_goals
data=[pl_not_scored,penalty_goals]
labels=['Penalties missed','Penalities scored']
plt.pie(data,labels=labels,autopct='%.0f%%')
plt.show()
```

```python
# unique positions
df['Position'].unique()
```

```python
# players from differt nation
np.size((df['Nationality'].unique()))
```

```python
# most players from which country
nation=df.groupby('Nationality').size().sort_values(ascending=False)
nation.head(10).plot(kind='bar')
```

```python
# club with maximum player in the squad
df['Club'].value_counts().nlargest(5).plot(kind='bar')
```

```python
# club with maximum player in the squad

df['Club'].value_counts().nsmallest(5).plot(kind='bar')
```

```python
# player based on age group

under20=df[df['Age']<=20]
age20_25=df[(df['Age']>20)&(df['Age']<=25)]
age25_30=df[(df['Age']>25)&(df['Age']<=30)]
above30=df[df['Age']>30]
```

```python
x = np.array([under20['Name'].count(),age20_25['Name'].count(),age25_30['Name'].count(),above30['Name'].count()])
mylabels=['<=20','20_25','25_30','>30']
plt.title('Total player with age',fontsize=20)
plt.pie(x, labels=mylabels, autopct='%.1f%%')
plt.show()
```

```python
# total under 20 age players in each club

player_under_20=df[df['Age']<20]
player_under_20['Club'].value_counts().plot(kind='bar')
```

```python
# avg age of each club

plt.figure(figsize=(12,6))
sns.boxplot(x='Club',y='Age',data=df)
plt.xticks(rotation=90)
```

```python
# total assists by each club

assist_by_club = pd.DataFrame(df.groupby(('Club'),as_index=False)['Assists'].sum())
ax = sns.barplot(x='Club',y='Assists',data=assist_by_club.sort_values(by="Assists"))
ax.set_xlabel("Club",fontsize=30)
ax.set_ylabel("Assists",fontsize=20)
plt.xticks(rotation=75)
plt.rcParams["figure.figsize"]=(20,8)
plt.title('plot of club vs total assist',fontsize = 20)
```

```python
# top 10 asssist

top_10_assist = df[['Name','Club','Assists','Matches']].nlargest(n=10,columns='Assists')
top_10_assist
```

```python
# total goals by each club

Goals_by_club = pd.DataFrame(df.groupby(('Club'),as_index=False)['Goals'].sum())
ax = sns.barplot(x='Club',y='Goals',data=Goals_by_club.sort_values(by="Goals"))
ax.set_xlabel("Club",fontsize=30)
ax.set_ylabel("Goals",fontsize=20)
plt.xticks(rotation=75)
plt.rcParams["figure.figsize"]=(20,8)
plt.title('plot of club vs Total Goals',fontsize = 20)
```

```python
# most goals by players

top_10_goals = df[['Name','Club','Goals','Matches']].nlargest(n=10,columns='Goals')
top_10_goals
```

```python
# pie chart - Goals with assist and without assist

plt.figure(figsize=(14,7))
assist = df['Assists'].sum()
data=[total_goals-assist,assist]
labels=['Goals without assist','Goals with assist']
plt.pie(data,labels=labels,autopct='%.0f%%')
plt.show()
```
